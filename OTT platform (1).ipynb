# Univariate Visualizations

```python
#Distribution of Watch Time
sns.histplot(df['watch_time_minutes'], kde=True, color= "orange")
plt.title("Watch Time Distribution")
plt.xlabel("Minutes")
plt.show()
```

- Most users watch for around 150 minutes.
- Theres a wide spread (from 5 to 300 minutes).

```python
# Watch Category Counts
df['watch_category'].value_counts().plot(kind='barh', color='darkcyan')
plt.title("Watch Category Counts")
plt.xlabel("Number of Users")
plt.ylabel("Watch Category")
plt.show()
```

- "Binge" may be the top category.
- This helps identify dominant viewing behaviors of the user

# Bivariate Analysis

```python
# Average Watch Time per Platform
df.groupby('platform')['watch_time_minutes'].mean().sort_values().plot(
    kind='barh',
    color='deeppink' )
plt.title("Avg Watch Time by Platform")
plt.xlabel("Minutes")
plt.show()
```

- Some platforms clearly encourage longer watch time.
- Netflix is on the highest .

```python
# Completion Rate by Genre

df.groupby('genre')['completion_rate'].mean().sort_values().plot(kind='barh', color='darksalmon')
plt.title("Avg Completion Rate by Genre")
plt.xlabel("Completion Rate")
plt.show()
```

- Most user are not tend to complete the Horror movies , they don't watch till end 
- Thriller and biography movies are the one which is completly watched by user 


```python
# Watch Time vs Watch Category
sns.barplot(data=df, x='watch_category', y='watch_time_minutes', estimator='mean', color='orchid')
plt.title("Average Watch Time by Category")
plt.show()
```

- Binge category likely has the highest average watch time, showing users spend more time when theyre binge-watching.
- Short Form or Quick Watch categories will show lower average watch time, as expected.
- The bar chart helps compare how much time people spend across different viewing habits.
- This info is helpful for content strategy  focus more on categories with higher engagement.

```python
location_platform = df.groupby(['location', 'platform'])['watch_time_minutes'].sum().reset_index()

plt.figure(figsize=(20, 7))
sns.barplot(data=location_platform, x='location', y='watch_time_minutes', hue='platform',palette='BuGn')

plt.title("Total Watch Time by Platform in Each location", fontsize=14)
plt.xlabel("location")
plt.ylabel("Total Watch Time")
```

**Total Watch Time by Platform in Each Location**
- This graph shows how much time users from different countries spent watching content on each OTT platform.
- India has the highest watch time overall, especially for YouTube and Zee5.
- In most countries (like the USA, UK, Germany, Brazil, Canada), Netflix and Prime Video dominate watch time.
- Hotstar is mainly significant in India, and to some extent in UAE and South Korea.
- SonyLiv and Zee5 have lower global presence, with watch time mainly contributed by Indian or Asian regions.
- YouTube performs well across all countries, indicating its universal reach and accessibility.

**Conclusion:**
- Netflix and Prime Video are globally dominant.
- YouTube is highly popular and widely used in all regions.
- Indian OTT platforms (Hotstar, SonyLiv, Zee5) have regional dominance, especially in India and UAE.

```python
genre_platform = df.groupby(['genre', 'platform'])['watch_time_minutes'].sum().reset_index()

plt.figure(figsize=(20, 7))
sns.barplot(data=genre_platform, x='genre', y='watch_time_minutes', hue='platform',palette='YlOrBr')

plt.title("Total Watch Time by Platform in Each Genre", fontsize=14)
plt.xlabel("Genre")
plt.ylabel("Total Watch Time (Minutes)")
```

**Total Watch Time by Platform in Each Genre**
- This graph displays platform popularity based on genres.
- YouTube leads in many genres like Documentary, Reality, Family, Music, and Comedy  genres suitable for shorter or informal content.
- Netflix and Prime Video dominate genres like Drama, Crime, Thriller, Romance, Mystery, and Sci-Fi  genres associated with longer-form storytelling.
- Zee5 and SonyLiv are more active in regional entertainment and some family and drama genres.
- Hotstar have more watch time in Sports, Family, and Drama, likely due to sports streaming and family-oriented Indian content.It can increse the watch-   time by attractinf=g people with similar intrest

**Conclusion:**
- YouTube excels in casual and quick-consumption genres.
- Netflix/Prime dominate in storytelling genres requiring binge-watching.
- Regional platforms tends to have more to cultural or local taste (e.g., drama and family genres).

```python
daily_watch = df.groupby('date')['watch_time_minutes'].sum()

plt.figure(figsize=(12, 5))
daily_watch.plot()
plt.title("Total Watch Time per Day")
plt.xlabel("Date")
plt.ylabel("Total Watch Time (minutes)")
plt.show()
```

- The plot shows the total number of watch-time minutes each day from Jan 2024 to 2025.
- The data has daily fluctuations, indicating that watch time varies day to day.
- There are several noticeable spikes, where total watch time crosses 3000+ minutes. These may indicate:
     - Special content releases 
     - Holidays or weekends when users watch more.
- The lowest dips go below 500 minutes, may be due to,
     -  Less engagement on specific days like weekdays, working days.

```python
sns.heatmap(df[['watch_time_minutes', 'completion_rate']].corr(), annot=True, cmap='YlGnBu')
plt.title("Correlation Between Watch Time and Completion Rate")
plt.show()
```

- There is not correlation between watch_time_minutes and completion_rate it is very close to zero(-0.0056)
- Even if the watch time increase it is not nessecary that the user will watch the show completely .
